# Main program : analysis of SI_clean.csv

```python
import pandas as pd
import numpy as np

%matplotlib tk
#%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np
import mplcursors

df_SI_clean = pd.read_csv("data/SI_clean.csv", delimiter=";", encoding="utf-8")
df_finlande_temp = pd.read_csv("data/finlande_temp.csv", delimiter=";", encoding="utf-8")

# Searching for non-int value, setting them to NaN
# TODO : set dtype to float
lastday_value = float(df_finlande_temp["janvier"][0])
for col in df_finlande_temp:
    cpt = 0
    for row in df_finlande_temp[col]:
        try:
            row = row.replace(",",".")
        except AttributeError as exc:
            # Row wasn't a str, it cant replace
            pass
        try:
            # If there is a difference between two days > 20°C, we assume that this is an error
            if abs(float(row) - lastday_value) > 19:
                df_finlande_temp.loc[cpt, col] = np.NaN
            else:
                lastday_value = float(row)
                df_finlande_temp.loc[cpt, col] = float(row)
        except ValueError:
            df_finlande_temp.loc[cpt, col] = np.NaN
        cpt+=1

# Columns set to float instead of "object"
df_finlande_temp = df_finlande_temp.astype(float)

# NaNs -> interpolated value
df_finlande_temp.interpolate(axis=0, inplace=True)

# Remove values for days that do not exist
months_length = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
for month_cpt, col  in enumerate(df_finlande_temp):
    for day_cpt, row in enumerate(df_finlande_temp[col]):
        if day_cpt+1 > months_length[month_cpt]:
            df_finlande_temp.loc[day_cpt, col]=np.NaN
            
df_finlande_temp
```

## Draw compare graph

```python
df_SI_clean_list = pd.DataFrame(df_SI_clean.values.T.ravel().tolist()).dropna()[0].tolist()
df_finlande_temp_list = pd.DataFrame(df_finlande_temp.values.T.ravel().tolist()).dropna()[0].tolist()
figzoom, axzoom = plt.subplots()
plt.plot(df_SI_clean_list)
plt.plot(df_finlande_temp_list)
plt.fill_between(range(0,365), df_SI_clean_list, df_finlande_temp_list, facecolor='yellow', alpha=0.5)
plt.show()
```
